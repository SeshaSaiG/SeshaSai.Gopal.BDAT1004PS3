# Step 3. Assign it to a variable called users

```python
# Assign variable for dataset
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', delimiter='|') 
```

# Step 4. Discover what is the mean age per occupation

```python
mean_age_per_occupation = users[['occupation', 'age']].groupby('occupation').mean()

mean_age_per_occupation
```

# Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
df = users[['gender', 'occupation']]

males = df.loc[df['gender'] == 'M']

group_by_occupations = df.groupby('occupation').count()
male_count_by_occupation = males.groupby('occupation').count()
male_ratio_per_occupation = male_count_by_occupation / group_by_occupations

male_ratio_per_occupation.sort_values(ascending=False, by='gender')
```

# Step 6. For each occupation, calculate the minimum and maximum ages

```python
df = users[['occupation', 'age']]

occ_min_age = df.groupby('occupation').min()

occ_max_age = df.groupby('occupation').max()

result = pd.concat([occ_max_age, occ_min_age], axis=1)

result
```

# Step 7. For each combination of occupation and sex, calculate the mean age

```python
df = users[['occupation', 'gender', 'age']]

mean_age_occupation_gender = df.groupby(['occupation', 'gender']).mean('age')

mean_age_occupation_gender
```

# Step 8. For each occupation present the percentage of women and men

```python
occ_percent = df.groupby('occupation')['gender'].value_counts(normalize=True)*100

occ_percent
```

# Create a line plot showing the number of marriages and divorces per capita in the U.S. between 1867 and 2014. Label both lines and show the legend. Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('us-marriages-divorces-1867-2014.csv')

df
```

```python
fig = plt.figure()

plt.rcParams["figure.figsize"] = [15, 10]


df.plot( 'Year', ['Marriages_per_1000','Divorces_per_1000'], label=['Marriages', 'Divorces'])


plt.title('Marriages and Divorces per capita in the U.S. between 1867 and 2014', color = 'k', fontsize = 22)


plt.xlabel('Year', color = 'k', fontsize = 17)
plt.ylabel('Capita', color = 'k', fontsize = 17)

plt.show()
```

# Create a vertical bar chart comparing the number of marriages and divorces per capita in the U.S. between 1900, 1950, and 2000. Don't forget to label your axes!

```python
filter_by_years = (df['Year'] == 1900) | (df['Year'] == 1950) | (df['Year'] == 2000) 

df_2 = df.loc[filter_by_years, ['Year','Marriages_per_1000','Divorces_per_1000']]

df_2
```

```python
df_2[{'Marriages_per_1000','Divorces_per_1000'}].plot.bar()

plt.title("Marriages and Divorces per capita in the U.S. between 1900, 1950 and 2000", color = 'm', fontsize = 22)

plt.xlabel('Year', color = 'm', fontsize = 20)
plt.ylabel('Capita', color = 'm', fontsize = 20)

plt.xticks((0, 1, 2), ("1900", "1950", "2000"), rotation = 0)
```
